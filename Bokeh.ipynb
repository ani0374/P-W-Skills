Q1. How can you create a Bokeh plot using Python code?

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers
output_file('test_html')
p=figure(title="test flower")
p.xaxis.axis_label='paltel length'
p.yaxis.axis_label='paltel width'
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
from bokeh.plotting import figure, show
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title="Example Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

p.line(x, y, legend_label="Line 1", line_width=2, line_color="blue")

p.circle(x, y, legend_label="Points", size=8, fill_color="red", line_color="black")

show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python

from bokeh.plotting import figure, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(
    title="Customized Bokeh Plot",
    x_axis_label="X-axis Label",
    y_axis_label="Y-axis Label",
   
)

p.line(x, y, legend_label="Line 1", line_width=2, line_color="blue")
p.circle(x, y, legend_label="Points", size=8, fill_color="red", line_color="black")

p.legend.title = "Legend Title"
p.legend.label_text_font_size = "12pt"

p.xaxis.major_label_text_color = "green"
p.yaxis.major_label_text_color = "purple"

p.background_fill_color = "lightgray"
p.grid.visible = True
p.grid.grid_line_color = "gray"

show(p)

```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

Bokeh Server is a component of the Bokeh library that allows you to create interactive data visualization applications in Python. It enables you to build web applications with live, interactive plots that can be updated in real-time as users interact with them. Bokeh Server is particularly useful when you want to create dynamic and interactive data dashboards or web-based data exploration tools.
